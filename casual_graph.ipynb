```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Duration
G = nx.DiGraph()

# Add edges based on the relationships you described
edges = [
    ("Outer Diameter", "Material ID"),
    ("Inner Diameter", "Material ID"),
    ("Height", "Material ID"),
    ("Weight", "Material ID"),
    ("Heat Treatment ID", "Material ID")
]

G.add_edges_from(edges)

# Add independent variables as isolated nodes
independent_vars = ["Product ID", "Workplace ID", "Duration Seconds", "productcategory_id"]
G.add_nodes_from(independent_vars)

# Draw the graph
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold", arrows=True)
plt.title("Manual Causal Graph for Duration")
plt.savefig("./casual_graphs/causal_graph_duration.png")
plt.show()
```

```python
# Stainless steel energy
G = nx.DiGraph()

# Add nodes
nodes = [
    "material_id", "outer_diameter", "height", "weight", 
    "heattreatment_id", "workplace_id", "product_id", "value"
]
G.add_nodes_from(nodes)

# Add edges based on relationships
edges = [
    ("material_id", "weight")
]

G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightgreen", font_size=10, font_weight="bold", arrows=True)
plt.title("Manual Causal Graph for Stainless Steel Energy")
plt.savefig("./casual_graphs/causal_graph_stainless_steel.png")
plt.show()
```
